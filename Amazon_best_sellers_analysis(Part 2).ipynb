```python
#Importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading in best sellers
best_sellers = pd.read_csv(R"C:\Users\hp\Documents\Amazon_best_sellers(2009-2021).csv")
```

```python
#displaying best sellers
best_sellers
```

```python
best_sellers.info()
```

```python
best_sellers.describe()
```

```python
# deleting unwanted columns
del best_sellers['Unnamed: 0']
```

```python
#deleting data for 2020 and 2021
list = [2020, 2021]

# drop rows from the above list
best_sellers = best_sellers[-best_sellers.year.isin(list) == True]
best_sellers
```

```python
#checking to confirm the columns were dropped
best_sellers['year'].unique()
```

```python
#checking the null values in the dataset
best_sellers.isnull().sum()
```

```python
#dropping the null values
best_sellers = best_sellers.dropna()
best_sellers
```

```python
#checking to see they were dropped
best_sellers.isnull().sum()
```

```python
#checking the dtypes
best_sellers.dtypes
```

```python
# convert rank to int and no_of_reviews to int
best_sellers['ranks'] = best_sellers['ranks'].astype('int64')
best_sellers['no_of_reviews'] = best_sellers['no_of_reviews'].str.replace(',','').astype('int64')
```

```python
df_1 = best_sellers.cover_type.value_counts().reset_index()
df_1
```

```python
#getting the distribution of cover pages 
plt.barh(df_1["cover_type"], df_1["count"], color='blue')
plt.title("cover type by number of occurence ")
plt.xlabel("No of occurence")
plt.ylabel("Cover type")
```

```python
#getting top 10 authors
df_2 = best_sellers.author.value_counts().reset_index().head(10)
```

```python
plt.barh(df_2["author"], df_2["count"], color='blue')
plt.title("Authors by number of appearance ")
plt.xlabel("No of appearance")
plt.ylabel("Authors ")
```

```python
#Top 10 books by appearance
df_3 =best_sellers.title.value_counts().head(10).reset_index()

df_3
```

```python
plt.barh(df_3["title"], df_3["count"], color='blue')
plt.title("Books by number of appearance ")
plt.xlabel("No of appearance")
plt.ylabel("Books")
```

```python
avg_price_and_year = best_sellers.groupby('year').price.mean().sort_values(ascending=False)
avg_price_and_year
```

```python
sns.barplot(x = avg_price_and_year.index, y= avg_price_and_year, color='r')
plt.ylabel('Price (dollars)', fontsize=15)
plt.xlabel('Year', fontsize=20)
plt.title('Year vs Average Price of Books', fontsize = 20);
```

```python
#top 10 authors by no_of_reviews
df_4= pd.DataFrame(best_sellers.groupby('author')['no_of_reviews'].mean().sort_values(ascending=False).iloc[:10].reset_index())
df_4
```

```python
plt.barh(df_4["author"], df_4["no_of_reviews"], color='red')
plt.title("Top 10 Author by Number of Reviews ")
plt.xlabel("No of reviews ")
plt.ylabel("Authors")
```

```python
#top 10 authors by the price pof their books
author_and_price = pd.DataFrame(best_sellers.groupby('author')['price'].mean().sort_values(ascending=False).iloc[:10].reset_index())
author_and_price
```

```python
#visualizing authors by mean price of their booksplt.rc('font', size=10)
plt.rc('xtick', labelsize=10) 
plt.rc('ytick', labelsize=10)

sns.barplot(data=author_and_price, x='price', y='author', color='b')
plt.ylabel('Author', fontsize=20)
plt.xlabel('Price ($)', fontsize=20)
plt.title('Best Selling Authors as Per Price', fontsize = 30)
plt.show()
```

```python
# Group by name, author, based on user rating.
review_wise = pd.DataFrame(best_sellers.groupby(['title', 'author'])['no_of_reviews'].mean())
review_wise.reset_index(inplace=True)

# sort values in descending order based on rating
review_wise = review_wise.sort_values(by='no_of_reviews', ascending=False)
review_wise.head(10)
```

```python
# visualizing the top 10 books based on user rating
#plt.figure(figsize=[8, 7])
#set label sizes for axes, xtick and ytick
plt.rc('font', size=30)
plt.rc('xtick', labelsize=30) 
plt.rc('ytick', labelsize=30)

sns.barplot(data=review_wise.head(10), x='no_of_reviews', y='title', color='b')
plt.ylabel('Book', fontsize=25)
plt.xlabel('Review', fontsize=25)
plt.title('Most Reviewed Books', fontsize = 30)
plt.show()
```
